```python
#toolkit for displaying cohomology generators 

def ordered_forms(generators: iter, deg: int):
    if not 1 <= deg < len(generators):
        print('degree not allowed')
        pass
    generators = list(set(itertools.combinations(generators, deg)))
    return sorted(generators)


def extract_vector(element):  #of the form Chain(1:(0, 1, 0, 0))
    element = str(element)
    gen = element[9:len(element) - 2]
    gen=gen.replace(', ', '').replace('-', '').replace('1/2','1').replace('3/2','1').replace('1/3','1').replace('2/3','1')
    vector = [int(gen[i]) for i in range(len(gen))]
    return vector

def translate_to_forms(homology_gens: list, deg: int, generators: iter):
    """
    return list of lists which are wedge products of generating forms. list(x^y^z, y^z^w,...)
    """
    homology_gens = homology_gens[0]

    generating_forms = []
    for n in range(len(homology_gens)):
        # vector = extract part of a string
        vector = extract_vector(homology_gens[n][1])
        generating_forms.append([])
        for i in range(len(vector)):
            if vector[i] != 0:
                generating_forms[n].append(ordered_forms(generators, deg)[i])
    return generating_forms


def cohomology_deg(cohomology: dict, deg: int):
    return [cohomology[key] for key in cohomology.keys() if key == deg]

def generate_cohomology(lie_algebra, deg):
    generators = list(lie_algebra.gens())
    if deg == 0 or deg >= len(generators):
        print('degree not allowed')
        return
    cochain = lie_algebra.chevalley_eilenberg_complex(dual=True)
    cohomology = cochain.homology(generators=True)
    cohomologies = [cohomology_deg(cohomology, n) for n in range(len(cohomology.keys()))]
    list_of_generators= translate_to_forms(cohomologies[deg], deg=deg, generators=generators)
    generators_readable=[]
    for i in range(len(list_of_generators)):
        tuple = (list_of_generators[i][0])
        string = str(tuple).replace(', ', '^').replace('(', '').replace(')', '').replace(',', '')
#         if len(string) == 5:  # forms of deg 1
#             string = string.replace(',', '')
        generators_readable.append(string)
    return generators_readable

```

```python
#toolkit for operating between forms and their string representations. 
#linearize is for generalizing functions to linear combinations

def to_form(form_string):
    return linearize(form_string, __to_form__, dic=dic)
    
    
def __to_form__(form_string, dic=dic):
    #works on basis
    if form_string=='0':
        return 0
    form_string=form_string.replace('^','')
    n=len(form_string)/2
    f=U.diff_form(n,'f')
    if n==1:
        i=int(form_string[1])-1
        f[i]=1
    if n>=2:
        vector=np.zeros(n)
        for i in range(n):
            vector[i]=int(form_string[i*2+1])-1
        vector=vector.astype(int)
        exec('f'+str(list(vector))+'=1')       
    return f


def to_string(form):
    if type(form)==str:
        form=simplify(form)
        'already a form string'
        return form
    try:
        if form==0:
            return '0'
    except StopIteration:
        return '0'    
    form_string= str(form.display())
    return simplify(form_string)


def simplify(form_string):
    form_string = form_string.replace('^','').replace('/\\','').replace('d','')
    try:
        return form_string.split(' = ')[1]
    except:
        #if form_string.split('-')[1][0]==' ':
        form_string = form_string.replace('- x','-1 x')
        return form_string

def linearize(form, nform_method, dic=dic):
    form_string=to_string(form) #form_string = '-2 x1 + x2'
    if form_string == '0':
        return 0
    elif is_1_form(form_string):
        if len(form_string)==2:
            return nform_method(form_string, dic)
        else:
            split=form_string.split(' ')
            result=float(split[0])*nform_method(split[1],dic)  
    else:
        split=form_string.split(' ')
        result=0
        for i in range(len(split)):
            if split[i]==0 or split[i]=='+':
                continue
            elif split[i]=='-' and is_constant(split[i+1]):
                split[i+1]=(-1)*float(Fraction(split[i+1]))
                continue
            elif is_constant(split[i]):
                split[i]=float(Fraction(split[i]))
                result += split[i]*nform_method(split[i+1],dic)
                split[i+1]=0
            else:
                result+= nform_method(split[i],dic)
    try:
        trial=result.display()
        return result
    except AttributeError:
        return 0

    
def is_constant(form_string):
    form_string=str(form_string)
    if 'x' in form_string:
        return False
    else:
        return True
    
def is_1_form(form_string):
    if form_string.count('x')==1:
        return True
    else:
        return False
    
```

```python
#basic operators on differential forms 

def __proj__(form_string, dic=dic):
    form=to_form(form_string)
    if d(form,dic)==0 and d_star(form,dic)==0:
        return form
    else:
        return 0
    
def proj(form, dic=dic):
    return linearize(form, __proj__, dic)
    
def pow(form, p):
    result=form
    if p==0:
        return 1
    elif p==1 or form == 0:
        return result
    else:
        for i in range(p-1):
            result= __wedge__(result,form)
    return result

def __d__(form_string, dic):
    #TODO the one with two gens in structure constants
    d=U.diff_form(2,'d')
    for structure_const in dic.values():
        if list(structure_const.keys())[0]==form_string:
            keys=[]
            for key, value in dic.items():
                if value == structure_const:
                    i=int(key[0][1])-1
                    j=int(key[1][1])-1
                    d[i,j]=list(structure_const.values())[0]
    try:
        trial=d.display()
        return d
    except ValueError:
        return 0

def d_of_nform(form_string,dic=dic):
    if form_string==0:
        return 0
    elif is_1_form(form_string):
        return __d__(form_string,dic)
    else:
        form1_string=form_string[:2]
        form2_string=form_string[2:]
        return __wedge__(d(to_form(form1_string),dic),to_form(form2_string)) - __wedge__(to_form(form1_string),d(to_form(form2_string),dic))

def d(form,dic=dic):
    return linearize(form, d_of_nform, dic)
            
def __wedge__(form1, form2):
    if form1==0 or form2==0:
        return 0
    elif form2 ==1:
        return form1
    elif form1==1:
        return form2
    else:
        return form1.wedge(form2)
    
def __star__(form_string, dic=dic):
    #works for nforms from basis
    if form_string=='0':
        return 0
    return psi_base(phi_inv_base(to_form(form_string)))

def star(form, dic=dic):
    return linearize(form, __star__, dic=dic)

def d_star(form, dic=dic):
    form_string=to_string(form)                  
    dim=len(form_string)/2
    return (-1)**dim*star(d(star(form),dic))
```

Implementing the star operator for contact manifold:
$$\phi:X^*(M) \to \Omega^*(M)$$
$$\phi^{-1}:\Omega^*(M)\to X^*(M)$$
$$\psi:X^*(M) \to \Omega^{2n-*}(M) $$
$$\star :\Omega^*(M) \to X^*(M) \to \Omega^{2n-*}(M), \; \star \beta =\psi \circ \phi^{-1}(\beta)$$


```python
def __phi__(v, dic=dic):
    contact=to_form('x7')
    deta=d(contact,dic)
    return v.interior_product(deta)

def create_phi_map_dict(dic=dic):
    #TODO fix the -1's
    phi_map={}
    for i in range(7):
        t=np.zeros(7)
        t[i]=1
        v=U.vector_field('v')
        v[:]=tuple(t)
        image=to_string(__phi__(v, dic))        
        if '-' in image:
            t[i]=-1*t[i]
            phi_map[tuple(t)]=image.replace('-','')
        else:
            phi_map[tuple(t)]=image 
    return phi_map

def invert_map(map_dictionary):
    return {v: k for k, v in map_dictionary.items()}
```

```python
def split_to_1forms(base_form):
    base_form=to_string(base_form)
    return [base_form[2*i:(2*i+2)] for i in range(len(base_form)/2)]

def __phi_inv__(form_string, dic=dic):
    phi_map=create_phi_map_dict(dic)
    inv_phi_map = invert_map(phi_map)
    #I have to map 'x7' to sth, but it's not used in B.
    inv_phi_map['x7'] = inv_phi_map['0']
    inv_phi_map.pop('0', None)
    print(inv_phi_map)
    return inv_phi_map[form_string]

def phi_inv_base(form):
    #works on base forms
    split=split_to_1forms(form)
    return [__phi_inv__(one_form) for one_form in split]

def compute_volume_form(dic=dic):
    contact=to_form('x7')
    deta=d(contact,dic)
    return deta.wedge(deta).wedge(deta)

def to_vector_field(vf_tuple):
    v=U.vector_field('v')
    v[:]=vf_tuple
    return v

def psi_base(list_vf, dic=dic):
    #returns a form
    list_vf=list(map(to_vector_field, list_vf))
    volume_form=compute_volume_form(dic)
    result=list_vf[::-1][0].interior_product(volume_form)
    n=len(list_vf)
    if n==1:
        return result
    for i in range(1,len(list_vf)):
        result=list_vf[::-1][0].interior_product(result)
    return result

```
